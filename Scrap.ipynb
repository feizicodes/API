```python
import requests
from bs4 import BeautifulSoup

url = "https://u.today/bitcoin-news"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Jeder Artikel ist innerhalb eines news__item-Blocks
articles = soup.find_all('div', class_='news__item')

for article in articles:
    title_tag = article.find('div', class_='news__item-title')
    link_tag = article.find('a', class_='news__item-body')
    date_tag = article.find('div', class_='humble')

    if title_tag and link_tag and date_tag:
        title = title_tag.get_text(strip=True)
        link = link_tag['href']
        date = date_tag.get_text(strip=True)

        print(f"Date: {date}")
        print(f"Title: {title}")
        print(f"Link: https://u.today{link}")
        print("------------")
```

```python
import requests
from bs4 import BeautifulSoup
import sqlite3

# Schritt 1: HTML von der Seite abrufen
url = "https://u.today/bitcoin-news"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Schritt 2: Verbindung zur SQLite-Datenbank aufbauen (oder neue Datei anlegen)
conn = sqlite3.connect("bitcoin_news.db")
cur = conn.cursor()

# Schritt 3: Tabelle anlegen, falls noch nicht vorhanden
cur.execute('''
    CREATE TABLE IF NOT EXISTS articles (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        date TEXT,
        title TEXT,
        link TEXT
    )
''')

# Schritt 4: Artikel-Daten extrahieren und speichern
articles = soup.find_all('div', class_='news__item')

for article in articles:
    title_tag = article.find('div', class_='news__item-title')
    link_tag = article.find('a', class_='news__item-body')
    date_tag = article.find('div', class_='humble')

    if title_tag and link_tag and date_tag:
        title = title_tag.get_text(strip=True)
        link = "https://u.today" + link_tag['href']
        date = date_tag.get_text(strip=True)

        # In Datenbank einfÃ¼gen
        cur.execute('''
            INSERT INTO articles (date, title, link)
            VALUES (?, ?, ?)
        ''', (date, title, link))

# Schritt 5: Ã„nderungen speichern und Verbindung schlieÃŸen
conn.commit()
conn.close()

print("âœ… Daten wurden erfolgreich in die Datenbank gespeichert.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# HTML abrufen
url = "https://u.today/bitcoin-news"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# CSV-Datei erstellen
with open("bitcoin_articles.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Date", "Title", "URL"])  # Header-Zeile

    # Artikel finden
    articles = soup.find_all('div', class_='news__item')

    for article in articles:
        title_tag = article.find('div', class_='news__item-title')
        link_tag = article.find('a', class_='news__item-body')
        date_tag = article.find('div', class_='humble')

        if title_tag and link_tag and date_tag:
            title = title_tag.get_text(strip=True)
            date = date_tag.get_text(strip=True)
            link = "https://u.today" + link_tag['href']

            # In CSV-Datei schreiben
            writer.writerow([date, title, link])

            # Optional: Kontrollausgabe
            print(f"{date}")
            print(f"{title}")
            print(f"{link}")
            print("------------")

print("âœ… Artikel erfolgreich in bitcoin_articles.csv gespeichert.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://u.today/bitcoin-news"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

articles = soup.find_all('div', class_='news__item')

with open("bitcoin_articles.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Date", "Title", "URL"])

    for article in articles:
        title = article.find('div', class_='news__item-title')
        link = article.find('a', class_='news__item-body')
        date = article.find('div', class_='humble')

        if title and link and date:
            writer.writerow([
                date.text.strip(),
                title.text.strip(),
                "https://u.today" + link['href']
            ])
```
